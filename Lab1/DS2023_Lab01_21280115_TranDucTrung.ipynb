```python
def count_char_freq(input_string):
    pass
    # Your code starts from here
    freq={}
    for char in input_string:
      if(char in freq): freq[char]+=1
      else: freq[char]=1
    return freq
    # Your code ends here

string = "abcaccdefeef"
char_freq = count_char_freq(string)
char_freq
```
